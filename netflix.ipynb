```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("netflix.csv")
netflix = pd.DataFrame(data)
```

```python
netflix
```

```python
netflix.dtypes
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')
```

```python
netflix.dtypes
```

```python
print(netflix['director'].isnull().sum()) # count missing values
print(netflix['cast'].isnull().sum()) 
print(netflix['country'].isnull().sum()) 
print(netflix['date_added'].isnull().sum()) 
print(netflix['rating'].isnull().sum()) 
```

```python
netflix['director'].fillna("unknown", inplace=True)
netflix['cast'].fillna("unknown", inplace=True)
netflix['country'].fillna("unknown", inplace=True)
netflix['date_added'] = netflix.apply(lambda x: pd.to_datetime(str(x['release_year']) + '-01-01') 
                                      if pd.isnull(x['date_added']) else x['date_added'], axis=1)

netflix['rating'].fillna("not available", inplace=True)
```

```python
print(netflix['director'].isnull().sum()) # count missing values
print(netflix['cast'].isnull().sum()) 
print(netflix['country'].isnull().sum()) 
print(netflix['date_added'].isnull().sum()) 
print(netflix['rating'].isnull().sum()) 
```
